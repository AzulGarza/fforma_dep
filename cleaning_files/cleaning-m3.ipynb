```python
import pandas as pd
import numpy as np
```

```python
year_data = pd.read_excel('data/data_m3/M3C.xls', sheet_name="M3Year")
quar_data = pd.read_excel('data/data_m3/M3C.xls', sheet_name="M3Quart")
month_data = pd.read_excel('data/data_m3/M3C.xls', sheet_name="M3Month")
other_data = pd.read_excel('data/data_m3/M3C.xls', sheet_name="M3Other")
```

```python
max_nrows = year_data.shape[1]
```

```python
year_df = []
nyear_test = 6
for idx, row in year_data.iterrows():
    id_serie = 'Y' + str(idx+1)
    n_size = row['N']
    year = row['Starting Year']
    serie = row.loc[list(np.arange(n_size) + 1)].dropna().to_numpy()
    len_serie = len(serie)
    index_serie = pd.date_range(str(year), periods=len_serie, freq='Y')
    category = row['Category']
    serie_df = pd.DataFrame({'id': id_serie,'fecha': index_serie, 'value': serie,  'freq': 'Y', 'category': category})
    #type of data
    type_data = np.array(['train']*len_serie)
    type_data[(len_serie-nyear_test):] = 'test'
    serie_df['type'] = type_data
    
    year_df.append(serie_df)
year_df = pd.concat(year_df)
```

```python
quar_df = []
nquar_test = 8
for idx, row in quar_data.iterrows():
    freq = 'Q'
    id_serie = 'Q' + str(idx+1)
    n_size = row['N']
    year = row['Starting Year']
    quar = row['Starting Quarter']
    if quar>4:
        quar = 4 #Error in data
    serie = row.loc[list(np.arange(n_size) + 1)].dropna().to_numpy()
    len_serie = len(serie)
    index_serie = pd.date_range(str(year)+'-'+str(quar*3), periods=len_serie, freq=freq)
    category = row['Category']
    serie_df = pd.DataFrame({'id': id_serie,'fecha': index_serie, 'value': serie,  'freq': freq, 'category': category})
    
    #type of data
    type_data = np.array(['train']*len_serie)
    type_data[(len_serie-nquar_test):] = 'test'
    serie_df['type'] = type_data
    
    quar_df.append(serie_df)
quar_df = pd.concat(quar_df)
```

```python
quar_df.tail(10)
```

```python
month_df = []
nmonth_test = 18
for idx, row in month_data.iterrows():
    freq = 'M'
    id_serie = 'M' + str(idx+1)
    n_size = row['N']
    year = row['Starting Year']
    mon = row['Starting Month']
    serie = row.loc[list(np.arange(n_size) + 1)].dropna().to_numpy()
    len_serie = len(serie)
    if year > 1900:
        index_serie = pd.date_range(str(year)+'-'+str(mon), periods=len_serie, freq=freq)
    else:
        index_serie = np.arange(len_serie) + 1
    category = row['Category']
    serie_df = pd.DataFrame({'id': id_serie,'fecha': index_serie, 'value': serie,  'freq': freq, 'category': category})
    
    #type of data
    type_data = np.array(['train']*len_serie)
    type_data[(len_serie-nmonth_test):] = 'test'
    serie_df['type'] = type_data
    
    month_df.append(serie_df)
month_df = pd.concat(month_df)
```

```python
month_df.tail(20)
```

```python
other_df = [] 
nother_test = 8
for idx, row in other_data.iterrows():
    freq = 'Other'
    id_serie = freq + str(idx+1)
    n_size = row['N']
    #year = row['Starting Year']
    #mon = row['Starting Month']
    serie = row.loc[list(np.arange(n_size) + 1)].dropna().to_numpy()
    len_serie = len(serie)
    #if year > 1900:
    #    index_serie = pd.date_range(str(year), periods=len_serie, freq=freq)
    #else:
    index_serie = np.arange(len_serie) + 1
    category = row['Category']
    serie_df = pd.DataFrame({'id': id_serie,'fecha': index_serie, 'value': serie,  'freq': freq, 'category': category})
    other_df.append(serie_df)
    
    #type of data
    type_data = np.array(['train']*len_serie)
    type_data[(len_serie-nother_test):] = 'test'
    serie_df['type'] = type_data
    
other_df = pd.concat(other_df)
```

```python
other_df.tail(10)
```

```python
total = pd.concat([year_df, quar_df, month_df, other_df]).reset_index(drop=True)
```

```python
total.to_csv('data/data_m3/dataM3.csv')
```
